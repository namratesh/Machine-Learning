```python
import pandas as pd 
```

* Seris : is a one-dimensional labeled array capable  of holding any data type(int,float,number,python object)
        the axis labels are collectively referred to as the index.
        

```python
series = pd.Series([1,2,3,'r',4])
series
```

* DataFrame: is two dimensional array labeld data structure with columns of different types.

```python
column = ['name', 'age', 'gender']
user1 = pd.DataFrame([ ['Andrew', 45, 'M'],
                     ['Scarlett', 32, 'F']],
                    columns = column)
user1
```

```python
column = ['name', 'age', 'gender']
user2 = pd.DataFrame([ ['Barry', 25, 'M'],
                     ['Arrow', 27, 'M']],
                    columns = column)
user2
```

* Concatenate Dataframe

```python
users = pd.concat([user1,user2])
users
```
