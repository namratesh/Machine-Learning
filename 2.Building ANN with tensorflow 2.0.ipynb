```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Flatten, Dense
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('churn.csv')
df.head()
```

```python
df.info()
```

```python
X = df.drop(labels = ['CustomerId', 'Surname', 'RowNumber', 'Exited'], axis =1)
y = df['Exited']
```

```python
X.head()
```

```python
y.head()
```

 Ann only accept numerical value so we have to change all string values in numerical value

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label = LabelEncoder()
 
X['Geography'] = label.fit_transform(X['Geography'])

X['Gender'] = label.fit_transform(X['Gender'])

```

After encoding we got 1 for male and 0 for female in gender
and in Geography  france :0 , spain:1, 

```python
X = pd.get_dummies(X, drop_first = True, columns = ['Geography'])
```

```python
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

Feature Standrization : is used get same scale for every feature

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

# ANN

```python
model = Sequential()
model.add(Dense(X.shape[1], activation  = 'relu', input_dim = X.shape[1]))
model.add(Dense(128, activation  = 'relu'))
model.add(Dense(1, activation  = 'sigmoid'))
model.summary()
```

```python
model.compile(optimizer = 'adam', loss='binary_crossentropy',
             metrics = ['accuracy'])
```

convert y_train to numpy array beacuse it is in pandas data

```python
model.fit(X_train, y_train.to_numpy(), batch_size = 12,epochs = 10,verbose=1 )
```

if loss is not decreasing the your model can be overfit

```python
y_pred = model.predict_classes(X_test)
```

```python
y_pred
```

```python
model.evaluate(X_test, y_test.to_numpy())
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
confusion_matrix(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```
